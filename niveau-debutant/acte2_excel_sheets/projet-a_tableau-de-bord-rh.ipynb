## Partie 1 — Mise en forme et tableau structuré

### Étape 1.1 — Convertir en tableau Excel

1. Clique sur n'importe quelle cellule dans tes données (ex : A1)
2. `Insertion → Tableau` (ou `Ctrl + T`)
3. Vérifie que "Mon tableau comporte des en-têtes" est coché → OK
4. Renomme le tableau : onglet **Création de tableau** → nom du tableau : **`Employes`**

> ✅ *Le tableau structuré permet à Excel de reconnaître automatiquement les nouvelles lignes et de référencer les colonnes par nom dans les formules.*

---

### Étape 1.2 — Mise en forme

1. **Figer la première ligne** : `Affichage → Figer les volets → Figer la ligne supérieure`
2. **Ajuster les largeurs** : sélectionne toutes les colonnes (A à J) → double-clic sur le bord d'une colonne pour ajuster automatiquement
3. **Colonne Contrat** : applique un code couleur avec une mise en forme conditionnelle :
   - `CDI` → fond vert clair (`#d4edda`)
   - `CDD` → fond orange clair (`#fff3cd`)
   - `Stage` → fond bleu clair (`#cce5ff`)

> Comment : `Accueil → Mise en forme conditionnelle → Règle de mise en évidence → Texte qui contient`

---

### Étape 1.3 — Vérification

Dans une cellule vide (ex : L1), tape :
```excel
=NBVAL(Employes[Prénom])
```

**Résultat attendu : 35** — si tu obtiens un autre chiffre, vérifie qu'il n'y a pas de lignes vides.

### 2.3 Bloc C — Formules avancées (SOMME.SI.ENS / NB.SI.ENS)

À partir de la cellule **A22**, réponds à ces questions :

**Question 1 :** Quelle est la masse salariale des CDI dans le département Commercial ?
```excel
=SOMME.SI.ENS(Données!I2:I36; Données!E2:E36;"Commercial"; Données!H2:H36;"CDI")
```
→ **Résultat attendu : 2 600 000 FCFA**

**Question 2 :** Combien y a-t-il de femmes en CDI dans tout le cabinet ?
```excel
=NB.SI.ENS(Données!D2:D36;"F"; Données!H2:H36;"CDI")
```
→ **Résultat attendu : 10**

**Question 3 :** Quel est le salaire moyen des hommes dans le département IT ?
```excel
=MOYENNE.SI.ENS(Données!I2:I36; Données!D2:D36;"M"; Données!E2:E36;"IT")
```
→ **Résultat attendu : 643 333 FCFA** *(Ibrahim Soro + Oumar Dembélé + Issa Diallo)*

> 💡 *`SOMME.SI.ENS`, `NB.SI.ENS`, `MOYENNE.SI.ENS` acceptent autant de critères que tu veux. C'est la version multi-conditions des fonctions `*.SI` simples.*

## Partie 4 — Graphiques

À partir de tes TCD, crée deux graphiques. Place-les dans la feuille `TCD` ou dans une nouvelle feuille `Graphiques`.

---

### Graphique 1 — Effectif par département (barres horizontales)

1. Sélectionne la colonne `Département` et la colonne `Effectif` du TCD 1 (sans la ligne Total)
2. `Insertion → Graphique → Barres → Barres groupées horizontales`
3. **Personnalise :**
   - Titre : `Effectif par département — Bamba & Associés`
   - Couleur des barres : bleu (`#0969da`)
   - Affiche les étiquettes de données (valeurs sur chaque barre)
   - Supprime la légende (inutile pour un seul indicateur)
   - Trie les départements par effectif décroissant si possible

---

### Graphique 2 — Répartition CDI / CDD / Stage (camembert)

1. Crée un petit tableau récapitulatif :
   - CDI : 21
   - CDD : 9
   - Stage : 5
2. Sélectionne ce tableau → `Insertion → Graphique → Secteurs (camembert)`
3. **Personnalise :**
   - Titre : `Répartition par type de contrat`
   - Étiquettes : affiche le **pourcentage** sur chaque secteur
   - Couleurs : CDI = vert, CDD = orange, Stage = bleu
   - Ajoute une légende

> 💡 *Pour le camembert, fais un clic droit sur le graphique → `Mettre en forme les étiquettes de données` → coche `Pourcentage` et décoche `Valeur`.*
